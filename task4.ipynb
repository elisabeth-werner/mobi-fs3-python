**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical $3 \times 3$ Prewitt filters:

```python
def prewitt_h(img):
    result = numpy.zeros(img.shape)
    kernel_h = numpy.array([[-1, 0, 1],
                            [-1, 0, 1],
                            [-1, 0, 1]])
    for p in numpy.ndindex(img.shape):
        if p[0] < 1 or p [1] < 1 or p[0] >= img.shape[0] - 1 or p[1] >= img.shape[1] - 1:
            continue
        neighborhood_sum_weighted = 0
        for q in numpy.ndindex(3,3):
            i, j = p[0] - 1 + q[0], p[1] - 1 + q[1]
            if 0 <= i < img.shape[0] and 0 <= img.shape[1]:
                neighborhood_sum_weighted += img[i,j] * 1/6 * kernel_h[q]
        result[p] = neighborhood_sum_weighted
    return result
    
def prewitt_v(img):
    result = numpy.zeros(img.shape)
    kernel_v = numpy.array([[-1, -1, -1],
                            [0, 0, 0],
                            [1, 1, 1]])
    for p in numpy.ndindex(img.shape):
        if p[0] < 1 or p [1] < 1 or p[0] >= img.shape[0] - 1 or p[1] >= img.shape[1] - 1:
            continue
        neighborhood_sum_weighted = 0
        for q in numpy.ndindex(3,3):
            i, j = p[0] - 1 + q[0], p[1] - 1 + q[1]
            if 0 <= i < img.shape[0] and 0 <= img.shape[1]:
                neighborhood_sum_weighted += img[i,j] * 1/6 * kernel_v[q]
        result[p] = neighborhood_sum_weighted
    return result
```

**<a id="#">Task 4.1.3.</a>** Test your above solution using the functions `prewitt_h` and `prewitt_v` for the previously loaded image and show the result:

```python
lenaprewitt_h = prewitt_h(img.copy())
plt.imshow(lenaprewitt_h, "gray")
plt.colorbar()

lenaprewitt_v = prewitt_v(img.copy())
plt.imshow(lenaprewitt_v, "gray")
plt.colorbar()
```
